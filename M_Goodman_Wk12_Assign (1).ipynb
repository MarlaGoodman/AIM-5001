#### Book 2: The Name of the Wind (XML)

The Name of the Wind was written by Patrick Rothfuss. It centers around one main character, Kvothe, who is primarily in one city, Imre, over the course of the book. I created an XML file with this information: https://raw.githubusercontent.com/MarlaGoodman/AIM-5001/master/namewind.xml

```python
#import urllib.request 
import urllib.request
#import objectify from lxml library
from lxml import objectify
```

The next two code cells are from the Week 12 lecture notes:

```python
#read xml file 
path, headers = urllib.request.urlretrieve('https://raw.githubusercontent.com/MarlaGoodman/AIM-5001/master/namewind.xml')

# parse through file with objectify() and open()
parsed = objectify.parse(open(path))

# create 'root' to house XML root node 
root = parsed.getroot()
```

```python
# create empty list to house for loop data
data = []

#for every item in the XML's root of 'overview', extract
for elt in root.overview:
     # items get a new dict
    el_data = {}
    #for every child tag, add the key/pair to the dictionaty
    for child in elt.getchildren():
        el_data[child.tag] = child.pyval
    #append new dictionary to our new list
    data.append(el_data)
    
#do the same loop, but for the 'attribute' root this time
info=[]
for elt in root.attribute:
    el_data = {}
    for child in elt.getchildren():
        el_data[child.tag] = child.pyval
    info.append(el_data)
    
#combine two new lists
result=data+info
result

#convert 'result' to df
Name_of_the_Wind=pd.DataFrame(result)
Name_of_the_Wind
```

This dataframe is different than the Harry Potter dataframe, in that it did not condense all the information into one row. It gave the 'overview' items (title and author) one row. Then it gave each attribute (character and city) its own row. This resulted in NaNs within the dataframe.

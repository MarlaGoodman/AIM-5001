### Task 2: Transforming Original Data

Import pandas library:

```python
import pandas as pd
```

Read csv file into notebook:

```python
df=pd.read_csv('https://raw.githubusercontent.com/MarlaGoodman/AIM-5001/master/hw10data.csv')
```

```python
#show df
df
```

To clean up the data, I am dropping the middle row that contains only NaNs.

```python
#where the LA column has an 'NaN', drop that whole row
dropped_df=df.dropna(subset = ['Los Angeles'])
dropped_df
```

I am now filling in the first NaN with 'ALASKA':

```python
#fill in NA with 'ALASKA', limiting this change only to the first NA
df_edited=dropped_df.fillna('ALASKA', limit=1)
```

And also filling in the last NaN with 'AM WEST':

```python
#fill in the NA of the df_edited dataframe with 'AM WEST'
data=df_edited.fillna('AM WEST')
```

```python
#show data
data
```

To make the data look a bit cleaner, I am renaming the columns:

```python
#rename columns of 'data' dataframe
data.columns = ['Airline', 'Arrival','Los Angeles','Phoenix','San Diego','San Francisco','Seattle']
data
```

The above dataframe is not tidy. I want each observation to get its own row. An observation is a very specific set of variables pertaining to flights. 

For example, the first observation is 497 planes that are apart of Alaska airline that arrived on time to LA. Therefore, in one row, I want 'Alaska', 'on time', 'LA', and '497'. 

This would then have the new dataframe contain 4 columns: 'Airline','Arrival' (on time arrival or delayed arrival),'Destination', and 'Number of Planes' (# of Planes). I will use the melt function to merge the 5 city columns into one column, which I will name 'Destination'.

```python
#melt the data so it appears on long form
melted=pd.melt(data,['Airline','Arrival'])
```

```python
#rename columns within melted data
melted.columns=['Airline','Arrival','Destination','# of Planes']
```

```python
#show melted data
melted
```

Question 2. Which airline had the best overall on time performance?

First I will get all the on time rows from the melted data using the same lambda even index function. Then I will group the Airlines and use the sum function. 

```python
#pull out even indexes ('on time' rows)
on_time=melted.iloc[lambda x: x.index % 2 == 0]
on_time
```

```python
#group by 'Airline' and sum the numbers
on_time.groupby(by='Airline').sum()
```

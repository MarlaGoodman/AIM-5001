I'll start by importing necessary libraries and the chess tournament text file:

```python
#importing pandas, regex libraries
import pandas as pd
import re
```

```python
#reading the text file into a dataframe, using the delimiter '|'
df=pd.read_csv('https://raw.githubusercontent.com/MarlaGoodman/AIM-5001/master/tournamentinfo.txt',delimiter='|')
```

```python
#convert dataframe to a string called 'text' in order to use regex on it
text=df.to_string()
```

```python
#show text
text
```

### Task 1:

I will first get the names of the players and their total points out of the text, using regex to do so:

```python
#find in the text everything that matches some variation of
#any capital letters followed by any character (4X),
#followed by spaces, followed by a number, then a period, then another number
step1=re.findall('[A-Z]+.[A-Z]+.[A-Z]+.[A-Z]+\s+[0-9]\.[0-9]',text)
step1
```

I want to manipulate the results above to get the data more organized. So first, I will convert the results into a string so I can use regex on them. Then I will use a regex that removes the unneeded spaces.

```python
#use comprehension for loop to join the results above into a string
listToStr1 = ' '.join([str(elem) for elem in step1]) 
listToStr1
```

```python
#implemeting regex to remove whitespaces
#by using \S+ to match any string of non-whitespace characters
step2=re.findall('\S+', listToStr1)
step2
```

However, now the resuls above have every value separated by a comma. I need the names of a player not to have commas between them. So first, I will turn the results into a string without commas. Second, I will extract all the information according to the players and their corresponding points.

```python
#converting results above into a string, joining it by a space and not a comma
listToStr2 = ' '.join([str(elem) for elem in step2])
listToStr2
```

```python
#using regex to pull out from the string above:
#everything that matches some variation of
#any capital letters followed by any character (4X),
#followed by space(s), followed by a number, then a period, then another number
names=(re.findall('[A-Z]+.[A-Z]+.[A-Z]+.[A-Z]+\s+[0-9]\.[0-9]',listToStr2))
names
```

**It is important for me to note here that the 'names' result above leaves out two players: Jade Ge and Ben Li. These names were removed from the data the moment I started manipulating the 'text'. I am not sure why this is the case. Nothing about Jade or Ben's data is different - format wise - than all the other players. The first regex used in this notebook ([A-Z]+.[A-Z]+.[A-Z]+.[A-Z]+\s+[0-9]\.[0-9]) brought back, among other data points, Jared Ge...a name nearly identical to Jade Ge. Therefore, I did not see why the regex did not bring back Jared and Ben's information. I spent hours trying to resolve it, but eventually decided to move forward for the sake of the assignment and its due date. To make sure this issue does not impede the rest of the projet, I manually added in Jared and Ben's names and points into their appropriate index.**

```python
#inserting Jared's and Ben's names and points into the appropriate index in 'names'
names.insert(32, 'JADE GE         3.5')
names.insert(63, 'BEN LI          1.0')
names
```

Now I will clean up the results above so that Jade and Ben's spacing is not any different than the other players:

```python
#convert the above results to a string, joined by only spaces
players = ' '.join([str(elem) for elem in names])
#Match any string of non-whitespace characters
mid=re.findall('\S+', players)
#convert back to string so I can continue working
player = ' '.join([str(elem) for elem in mid])
```

Now that all players are accounted for, I want to separate the names from the points, creating two separate lists that will be used for later.

```python
#use regex to just pull out letters from the new string
#(any capital letters that have spaces, or spaces and a dash)
name=re.findall('[A-Z\s-]+',player)
name
```

```python
#use regex to just pull out numbers from the new string
#(find one number, followed by a period, followed by one number)
number=re.findall('[0-9]\.[0-9]',player)
number
```

### Task 2:

I will now get the information of where the player is from. I'll start off by pulling that information from the 'text', using regex to do so. If needed, I will clean up the data and its format.

In 'text', the location initials come after an '\n'. I will use a regex starting with that to pull out the location information.

```python
#\\n means bring back a literal '\n'
# match a \n, followed by space(s), followed by exactly 2 uppercase letters
place=re.findall('\\n\s+[A-Z]{2}',text)
place
```

I will now clean up results so that I no longer have the '\n':

```python
# converting results to a string with no commas to work on the data further
string_place = ' '.join([str(elem) for elem in place])
string_place

#from the string, match all uppercase letters
places=re.findall('[A-Z]+',string_place)
places
```

### Task 3:

I will now get each players pre-rating. I will use regex to pull that information out, and format it appropriately. 

```python
#match within 'text' wherever there is a colon, followed by space(s), followed by number(s)
pr=(re.findall('\:\s+[0-9]+',text))

#convert results of pr to a string for further cleanup of the data
pr2=str(pr)
pr2
```

```python
#within pr2, match all the numbers (numbers only)
rating=re.findall('[0-9]+',pr2)
rating
```

### Task 4:

Now that I have my lists of rating, places, name, and number, I will put them all together, separated by commas.

```python
#use zip function to take iterables, aggregate them, and return the result
#via a comprehension for loop
npnr=[list(l) for l in zip(name, places, number, rating)]
```

```python
#show final
npnr
```

### Task 5:

I will now put the results above into a pandas dataframe.

```python
# create data frame from npnr results above
#add column names
tournament_info = pd.DataFrame.from_records(npnr, columns = ['Player', 'State', 'Total Points', 'Pre-Rating'])
tournament_info
```

### Task 5:

Finally, I will generate a .CSV file from the dataframe above.

```python
# generate a CSV based the dataframe above
tournament_info.to_csv('tournament_info.csv', index = False)
```
